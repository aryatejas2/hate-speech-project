```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Tejas\Desktop\Capstone\DatasetsCleaned\tcc.csv")
```

```python
data = data[['comment_text','toxic','severe_toxic','obscene','threat','insult','identity_hate']]
data
```

```python
data['Class'] = ""
```

```python
for row in range(len(data)):
    if (data['toxic'][row] == 1 or data['severe_toxic'][row] == 1 
                or data['obscene'][row] == 1 or data['threat'][row] == 1
                or data['insult'][row] == 1 or data['identity_hate'][row] ==1):
            data['Class'][row] = "OFF"
    else:
        data['Class'][row] = "NOT"
```

```python
data = data[['comment_text','Class']]
```
